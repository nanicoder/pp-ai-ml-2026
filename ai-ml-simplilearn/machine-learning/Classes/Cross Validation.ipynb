```python
from sklearn.model_selection import cross_val_score
lr = LinearRegression()

scores = cross_val_score(lr,   # model that should be built on all the sets od data
                         cal_housing_data.drop(columns = ['MedHouseVal']) , # x --> prdictors
                         cal_housing_data['MedHouseVal'],  # y  --> label / dependent
                         cv = 5 ,  # number of splits
                         scoring = 'r2')
scores
```
