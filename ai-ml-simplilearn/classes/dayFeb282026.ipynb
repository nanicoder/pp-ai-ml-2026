```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
shippingdimenData = pd.read_csv('../data-sets/shippingdimen.csv')
marketfactData = pd.read_csv('../data-sets/marketfact.csv')
custdimenData = pd.read_csv('../data-sets/custdimen.csv')
proddimenData = pd.read_csv('../data-sets/proddimen.csv')
```

```python
marketfactProdData = pd.merge(marketfactData, proddimenData, on = 'Prod_id', how = 'outer')
marketfactProdData.to_csv('../data-sets/marketfactProdData.csv', index = False)
```

```python
#Combine the two dataframes using the common column 'cust_id'
marketfactProdCustData = pd.merge(marketfactProdData, custdimenData, on = 'Cust_id', how = 'outer')
marketfactProdCustData.to_csv('../data-sets/marketfactProdCustData.csv', index = False)
```

```python
marketfactProdCustShippingData = pd.merge(marketfactProdCustData, shippingdimenData, on = 'Ship_id', how = 'outer')
marketfactProdCustShippingData.to_csv('../data-sets/marketfactProdCustShippingData.csv', index = False)
```

```python
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
population = [7.3, 7.4, 7.5, 7.4, 7.7, 7.8, 7.7, 8.0, 8.1, 8.2]
```

```python
plt.plot(year, population)
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.title('World Population Growth')
plt.grid()
plt.show()
```

```python
plt.plot(year, population, color = 'green', marker = '*', linestyle = '--')
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.title('World Population Growth')
plt.grid()
plt.show()
```
