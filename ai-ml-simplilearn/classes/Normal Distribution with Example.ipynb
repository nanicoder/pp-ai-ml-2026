```python
# Step 1 -> you have a data and wants to calculate probabily for a value 

# Step 2 -> To ensure the data is normally distributed , if you plan to use PDF (formula for getting prob) of normal distribution to calcuate prob.
                # how to check the distribution -> grpahs - histograms, tests

#  for sake of understanding -> lets say our data is normally distributed
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
school_bag_weight_data = np.random.normal(loc = 6 , scale = 0.5 , size = 1000).round(2) # loc = mean , scale = standard deviation 
school_bag_weight_data
```

```python
plt.hist(school_bag_weight_data , bins = 30, density = True)
plt.show()
```

Question is - what is the probabilty of the school bag weighing more than 5.8 kg and less than 6.1 kg

Sol - We can use PDF directly and do all the calculations, but to make our life easier by avoiding calculation, we want to use the table where all prob. are given

- The problem is
    - the table is standard (https://math.arizona.edu/~jwatkins/normal-table.pdf) , as it not possible to have a table for all different scales of data
 
- So we convert our data to standard form ( it can be done for any data) . how -> by using Z score : (x-mu) / std

```python
mean = np.mean(school_bag_weight_data).round(2)
std = np.std(school_bag_weight_data).round(2)

print(mean , std)
```

```python
# convert data to z scores
data_pt_z_score = {}
z_scores = []

for i in school_bag_weight_data:
   data_pt_z_score[i] = round( ((i - mean) / std) , 2)
   z_scores.append(round( ((i - mean) / std) , 2)) 

data_pt_z_score

# as my data was normally distributed, hence now it converted to standard normal distribution , mean = 0, std = 1
```

```python
print( np.mean( z_scores ) )
print( np.std ( z_scores ) )
```

```python
print(min(data_pt_z_score.values()))
max((data_pt_z_score.values()))
```

```python
# z_score for mean
z_score_for_mean = (6.02 - mean) / std
z_score_for_mean
```

```python
# convert 5.8kg and 6.1kg to z score ,  as our question is for 5.8 to 6.1 kg
z_score_for_lower = (5.8 - mean) / std
z_score_for_lower
```

```python
# convert 5.8kg and 6.1kg to z score ,  as our question is for 5.8 to 6.1 kg
z_score_for_upper = (6.1 - mean) / std
z_score_for_upper
```
