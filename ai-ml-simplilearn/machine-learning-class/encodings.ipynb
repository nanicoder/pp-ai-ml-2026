## Ordinal encoding

```python
test_df = pd.DataFrame({'size': ['small', 'medium', 'large', 'medium', 'small', 'large'],
                        'quality' : ['low' , 'low' , 'medium', 'low', 'high', 'high']})
```

```python
test_df
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = [   ['small', 'medium', 'large']  , ['low', 'medium', 'high']  ])
```

```python
oe.fit_transform(test_df)
```

```python
pd.DataFrame(data = oe.fit_transform(test_df) , columns = test_df.columns)
```
